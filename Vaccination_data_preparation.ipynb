### Get Twitter posts

```python
df_twitter = df.loc[df["Джерело"]=="twitter.com"]
```

```python
for col in df_twitter.columns:
    print(col+":",len(df_twitter[col].unique()))
```

```python
df_twitter_dropped = df_twitter[["Дата","Час","Збережено","Дайджест тексту","Тип посту","Тональність","Автор",
                                "Підписники","Сума всіх реакцій","Лайки","Коментарі","Репости"]]
```

```python
df_twitter_cleared = df_twitter_dropped.fillna(0) 
```

### Get Facebook posts

```python
df_fb = df.loc[df["Джерело"]=="facebook.com"]
```

```python
for col in df_fb.columns:
    print(col+":",len(df_twitter[col].unique()))
```

```python
dropping = ["URL","Опрацьоване","Призначено","Рейтинг","Перегляди","Dislikes","Нотатки","Тип джерела",
           "Профіль місця публікації","Джерело","Профіль","Love","Haha","Wow","Sad","Angry","URL зображення",
           "Вік"]

df_fb_dropped = df_fb.drop(dropping,axis=1)
```

```python
df_fb_dropped
```

```python
df_fb_cleared = df_fb_dropped.fillna(0)
```

```python
df_fb_cleared.to_excel("facebook_vaccination.xlsx")
```
